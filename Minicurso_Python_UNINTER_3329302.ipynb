#### Utilizando condicionais (If, Elif e Else)

```python
# Testando uma condicional simples
first_var = 1
second_var = 2
if first_var == second_var:
  print(f'Os valores {first_var} e {second_var} são iguais!')
else:
  # print(f'As variáveis {first_var} e {second_var} têm valores distintos!')
  print('As variáveis {} e {} têm valores distintos!'.format(first_var, second_var))

# Testando uma condicional com elif
third_var = 1

if first_var == second_var:
  print(f'Os valores {first_var} e {second_var} são iguais!')
elif first_var == third_var:
  print(f'Os valores {first_var} e {third_var} são iguais!')
else:
  print("A variável 'first_var' é única!")
```

```python
# Testando uma condicional com textos
first_text = 'Olá, Pedro'
second_text = 'Tchau, Pedro'
third_text = 'Olá, Pedro'

if first_text == third_text:
  print('Textos 1 e 3 semelhantes!')
elif first_text == second_text:
  print('Textos 1 e 2 semelhantes!')
else:
  print('O texto 1 não tem par.')
```

#### Estruturas e Laços de Repetição

```python
# Utilizando "for"
for numero in range(0, 11, 1): # O index no Python se inicia no 0
  print(numero)
```

```python
# Mais exemplos práticos utilizando for:
for name in primeira_lista:
  print(f'O nome {name} possui {len(name)} letras.')
```

```python
# Unindo o laço FOR com condicionais (buscando números pares)
for num in range(1, 11, 1):
  if num % 2 != 0:
    print(f'O número {num} é impar.')
  else:
    print(f'O número {num} é par.')
```

```python
# Utilizando While
index = 10
while index > -1:
  print(index)
  index -= 1
```

```python
# While True
while True:
  print("olá")
```

#### Try e Except (com input)

```python
# Advinhando um número com as estruturas condicionais e laços de repetição
while True:
  try:
    num = int(input('Insira um número entre 0 e 10:'))  # Convertendo o input para inteiro
    if num > 10 or num < 0:
      print('Insira apenas números entre 0 e 10!')
      continue
    elif num != 5:
      print('Número incorreto. Tente novamente')
      continue
    else:
      print('Você acertou!!')
      break
  except ValueError:  # Captura o erro caso o usuário não insira um número
    print('Por favor, insira um número válido!')
```
